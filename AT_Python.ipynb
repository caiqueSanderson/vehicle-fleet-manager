# Questão 2 | Cadastro Veicular

Crie uma função que realize o cadastro do veículo, recebendo com passagem uma lista com o banco de dados dos veículos e retorne essa mesma lista com o novo valor de registro atualizado. O campo de disponibilidade deve ter valor -1 para quando o veículo estiver disponível e o índice do cliente que o locou, cado esteja indisponível.

```python
def cadastrar_veiculo(bd_veiculos):
  print('----- Cadastro Veículo -----')
  placa = input('Insira a placa do veiculo: ').upper()
  modelo = input('Insira o modelo do veiculo: ')
  categoria = input('Insira a categoria do veiculo: \n(Categorias: básico, intermediário e luxo) \n').lower()
  estado = input('Insira "s" para disponivel ou "n" para indisponível: ').lower()

  if estado == 's':
    estado = -1
  else:
    estado = input('Insira o ID do cliente que locou o veiculo:')

  bd_veiculos.append([placa, modelo, categoria, estado])
  return bd_veiculos
```

# Questão 4 | Busca veicular(Placa)

Crie uma função que realize a busca de um veículo a partir da sua placa. Ela deve receber como passagem a lista contendo o banco de dados de veículos e a placa, e retorne uma lista contendo o índice do veículo desejado.

```python
def buscar_veicular_placa(bd_veiculos, placa_procurada):
  veiculo_encontrado = []
  for indice, veiculo in enumerate(bd_veiculos):
    if placa_procurada == veiculo[0]:
      veiculo_encontrado.append(indice)
    elif indice == len(bd_veiculos):
      print('Placa não encontrada')
    else:
      continue
  return veiculo_encontrado
```

# Questão 5 | Busca veiculo(Disponibilidade)

Crie uma função que realize a busca por veículos, tendo como base o estado da locação. Ela deve receber como parâmetros de entrada a lista contendo o banco de dados dos veículos e um parâmetro booleano, que selecionará se a lista retornada será dos veículos disponíveis ou indisponíveis. O valor padrão para esse parâmetro deve ser disponivel=True. O retorno da função deve uma lista contendo os índices de todos os veículos selecionados.

```python
def buscar_veiculo_disponibilidade(bd_veiculos,disponibilidade=True):
  if disponibilidade == True:
    veiculos_disponiveis = []
    for indice,lista in enumerate(bd_veiculos):
      if lista[3] == -1:
        veiculos_disponiveis.append(indice)
    return veiculos_disponiveis

  else:
    veiculos_indisponiveis = []
    for indice,lista in enumerate(bd_veiculos):
      if lista[3] != -1:
        veiculos_indisponiveis.append(indice)
    return veiculos_indisponiveis
```

# Questão 7 | Busca Cliente

Crie uma função que faça a busca por cliente a partir do seu o nome ou o sobrenome. Deve receber como parâmetros a lista contendo o banco de dados dos clientes, o nome que deve ser localizado e um parâmetro booleano que informe se a busca deve ser realizada pelo nome ou sobrenome. O retorno deve ser uma lista contendo todos os índices onde houveram ocorrência do nome/sobrenome desejados.

```python
def buscar_cliente(bd_clientes,nome,sobrenome = False):
  registros_encontrados = []
  if sobrenome == False:
    for indice,lista in enumerate(bd_clientes):
      if lista[0][0] == nome:
        registros_encontrados.append(indice)
    return registros_encontrados

  else:
    for indice,lista in enumerate(bd_clientes):
      if lista[0][1] == nome:
        registros_encontrados.append(indice)
    return registros_encontrados
```

# Questão 9 | Alugar Veículo

Crie uma função que realize o aluguel do veículo. Ela deve receber como parâmetros as listas contendo os bancos de dados de veículos e clientes e retorná-los atualizados. A função deve solicitar a placa do veículo e avaliar se ele está disponível para locação. Caso não esteja, deve oferecer ao usuário a opção de inserir uma nova placa ou sair da opção. Caso o veículo esteja disponível, ela deve exibir os dados do veículo e pedir que o usuário confirme que está correto. Caso negativo, ofereça a opção de inserir uma nova placa ou sair. Se estiver correto, solicite o ID do cliente, faça a mesma avaliação e ofereça as mesmas opções apresentadas no caso dos veículos. Se tudo estiver correto, os campos de ‘disponibilidade do veículo’ e de ‘veículo locado’ devem ser atualizados nas respectivas bases de dados e um sumário da operação exibido na tela. A função deve retornar a duas listas atualizadas.

```python
def alugar_veiculo(bd_veiculos,bd_clientes):
  while True:
    solicita_placa = input('Informe a placa do veiculo: ')
    verifica_placa = buscar_veicular_placa(bd_veiculos, solicita_placa)

    for indice in verifica_placa:
      if bd_veiculos[indice][3] != -1:
        print('\nEsse veículo já está alugado. \n')
        nova_placa_ou_sair = int(input('Insira 1 para inserir uma nova placa ou 2 para sair: '))
        if nova_placa_ou_sair == 1:
          continue
        elif nova_placa_ou_sair == 2:
          return
      elif bd_veiculos[indice][3] == -1:
        print(f'Dados do veículo inserido: {bd_veiculos[indice]}')
        confirmacao_veiculo = input(f'Esse é o veículo desejado? \n(Sim ou Não) \n').lower()

        if confirmacao_veiculo == 'sim':
          while True:
            solicita_id = int(input('Informe o ID do cliente: '))
            if solicita_id < 0 or solicita_id - 1 >= len(bd_clientes):
              print("ID do cliente inválido.")
              continue
            else:
              print(f'Dados do cliente: {bd_clientes[solicita_id - 1]}')
              confirmacao_cliente = input(f'Esse é o cliente desejado? \n(Sim ou Não) \n').lower()

              if confirmacao_cliente == 'sim':
                bd_veiculos[indice][3] = solicita_id
                bd_clientes[solicita_id - 1].append(solicita_placa)
                print(f'\n---- Veículo alugado com sucesso ----\nVeículo: {bd_veiculos[indice]} \nCliente: {bd_clientes[solicita_id - 1]}')
                return bd_veiculos, bd_clientes
                break

              elif confirmacao_cliente == 'não':
                novo_id_ou_sair = int(input('Insira 1 para inserir um novo ID para buscar cliente ou 2 para sair: '))
                if novo_id_ou_sair == 1:
                  continue
                elif novo_id_ou_sair == 2:
                  return
        elif confirmacao_veiculo == 'não':
          nova_placa_ou_sair = int(input('Insira 1 para inserir uma nova placa ou 2 para sair: '))
        if nova_placa_ou_sair == 1:
          continue
        elif nova_placa_ou_sair == 2:
          return
```

# Questão 10 | Devolução Veículo

Crie uma função que faça a devolução do veículo, contabilizando o valor total da locação. Deve receber como parâmetros as listas de veículos e de clientes e retornar essas listas atualizadas. Deverá ser solicitado o nome ou o código do cliente. O processo de devolução deve desvincular o veículo do usuário, retornando-o para o estado ’disponível’ e remover o veículo da base do cliente.
```python
def devolucao_veiculo(bd_clientes,bd_veiculos):
  nome_ou_id = int(input('Insira 1 para procurar por nome ou 2 para procurar pelo ID do cliente: '))
  if nome_ou_id == 1:
    nome = input('Insira o nome do cliente: ')
    busca = buscar_cliente(bd_clientes,nome,sobrenome = False)
    for cliente in busca:
      print(cliente)

  elif nome_ou_id == 2:
    while True:
      id = int(input('Insira o ID do cliente: '))
      print(f'{bd_clientes[id - 1]}\n')
      confirmacao = input('Confirma que esse é o cliente procurado?').lower()
      if confirmacao == 'sim':
        placa = bd_clientes[id - 1].pop(-1) #Desvincula a placa do cliente e retorna a placa
        indice = buscar_veicular_placa(bd_veiculos, placa)[0] #O retorno da função é uma lista
        dados_veiculo = bd_veiculos[indice]
        dados_veiculo[-1] = -1

        numero_diarias = int(input('Insira a quantidade de diárias utilizadas: '))
        conta = 0
        categoria = dados_veiculo[2]
        if categoria == 'básico':
          conta = numero_diarias * 150
        elif categoria == 'intermediário':
          conta = numero_diarias * 250
        elif categoria == 'luxo':
          conta = numero_diarias * 400

        print(f'----- Total ----- \nO valor da conta é: R${conta},00')
        return

      elif confirmacao == 'não':
        continue
```
